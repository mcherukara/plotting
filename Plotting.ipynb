# Make nice figures

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
plt.style.use('seaborn-white')
#matplotlib.rc('xtick', labelsize=20) 
#matplotlib.rc('ytick', labelsize=20)
matplotlib.rc('font',family='Times New Roman')
matplotlib.rcParams['font.size'] = 20
```

# Load data

```python
histx=np.loadtxt('simplex_x.txt', skiprows=1)
histz=np.loadtxt('simplex_z.txt', skiprows=1)
print histx.shape, histz.shape
```

```python
xpts=np.linspace(0,histx.shape[0],histx.shape[0]+1)
zpts=np.linspace(0,histz.shape[0],histz.shape[0]+1)
```

```python
plt.plot(xpts[1:],histx[:,0], 'C3o-', label='X focus')
plt.xlabel('Iterations')
plt.ylabel('FWHM ($\mu$m)')
plt.tight_layout()
plt.grid()
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5))
plt.tick_params(which='both', width=2)
plt.tick_params(which='major', length=7)
plt.tick_params(which='minor', length=5)
#plt.savefig('figs/x_iters.png', dpi=200)
```

```python
plt.plot(zpts[1:],histz[:,0], 'C0o-')
plt.xlabel('Iterations')
plt.ylabel('FWHM ($\mu$m)')
plt.tight_layout()
#plt.savefig('figs/z_iters.png', dpi=200)
```

# Share axis

```python
f, axarr = plt.subplots(2, sharex=True, figsize=(8, 6))
axarr[0].plot(xpts[1:],histx[:,0], 'C3o-', label='X focus')
axarr[0].set(ylabel='FWHM ($\mu$m)')
axarr[0].grid()
axarr[1].plot(zpts[1:],histz[:,0], 'C0o-', label='Z focus')
plt.xlabel('Iterations')
plt.ylabel('FWHM ($\mu$m)')
plt.tight_layout()
plt.grid()
plt.tick_params(which='major', length=7, width=2)
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5))
```

```python
x,y,z=histx[:,2],histx[:,3], histx[:,0]
plt.tricontourf(x,y,z, 10, cmap='viridis')
plt.xlabel('Motor 1 (arb.)')
plt.ylabel('Motor 2 (arb.)')
cb=plt.colorbar()
cb.set_label("FWHM ($\mu$m)")
plt.tight_layout()
#plt.savefig('figs/x_map.png', dpi=200)
```

```python
x,y,z=histz[:,2],histz[:,3], histz[:,0]
plt.tricontourf(x,y,z, 10, cmap='viridis')
plt.xlabel('Motor 1 (arb.)')
plt.ylabel('Motor 2 (arb.)')
cb=plt.colorbar()
cb.set_label("FWHM ($\mu$m)")
plt.tight_layout()
#plt.savefig('figs/z_map.png', dpi=200)
```
